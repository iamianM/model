```python
import pandas as pd
from os import listdir
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
import mplfinance as fplt
from os import path
import os
```

```python
plt.style.use('ggplot')
```

```python
PATH = './dukascopy'
```

```python
for file in [f for f in listdir(PATH) if f.endswith('.csv')]:
    symbol = file.split('.')[0]
    print(symbol)
    df = pd.read_csv(f'{PATH}/{file}')
    df = df[df['Volume'] > 0]
    peaks, peak_prop = find_peaks(df['Close'], width=1, prominence=(0.02, None))
    valleys, valley_prop = find_peaks(df['Close']*(-1), width=2.5, prominence=(0.05, None), plateau_size=(0,1))
    df['label'] = ['Sell' if idx in peaks and df.iloc[idx]['Local time'].split(' ')[1] >= '07:15:00' and df.iloc[idx]['Local time'].split(' ')[1] <= '13:00:00' else 'Buy' if idx in valleys and df.iloc[idx]['Local time'].split(' ')[1] >= '07:15:00' and df.iloc[idx]['Local time'].split(' ')[1] <= '12:00:00' else 'Hold' for idx in range(len(df))]
    df['Local time'] = pd.to_datetime(df['Local time'])
    df = df.set_index('Local time')
    df.to_csv(f'{PATH}/stock_data/{symbol}.csv')
    
    dates = [idx.isoformat().replace(':', '-') for idx in df.index]
    for i in range(len(df)):
        if i %1000 == 0:
            print(i)
        if i-44 >= 0:
            date = df.iloc[i:i+1].index[0].isoformat().replace(':', '-')
#             image_file = f'{PATH}/stock_graphs/all/{symbol}_{dates[i]}.png'
            image_file = f'{PATH}/stock_graphs/all/{symbol}_{dates[i]}.png'
            if path.exists(image_file):
                continue
                
            mc = fplt.make_marketcolors(
                up='tab:blue',down='tab:red',
                edge='black',
                wick={'up':'blue','down':'red'},
                volume='green', alpha=1.0
            )
            s  = fplt.make_mpf_style(marketcolors=mc, mavcolors=["yellow","orange",'brown'])

            fig = fplt.plot(
                df.iloc[i-44:i+1],
                type='candle',
                style=s,
                volume=True,
                axisoff=True,
                mav=(5,8,13),
                figsize=(3.2,3.2),
                xlim=(12.5, 44.5),
                savefig=image_file,
                tight_layout=True
            )
```

```python
def check_folder_exists(folder_location):
    if not os.path.exists(folder_location):
        os.makedirs(folder_location)
```

```python
for file in [f for f in listdir('./daily_stock_data/') if f.endswith('.csv')]:
    symbol = file.split('_')[0]
    print(symbol)
    df = pd.read_csv(f'./daily_stock_data/{file}')
    df.columns = ['Local time' if c == 'time' or c == 'timestamp' else c.capitalize() if c != 'label' else c for c in df.columns]
    df = df[df['Volume'] > 0]
    peaks, peak_prop = find_peaks(df['Close'], width=1, prominence=(0.02, None))
    valleys, valley_prop = find_peaks(df['Close']*(-1), width=2.5, prominence=(0.05, None), plateau_size=(0,1))
    df['label'] = ['Sell' if idx in peaks and df.iloc[idx]['Local time'].split(' ')[1] >= '10:15:00' and df.iloc[idx]['Local time'].split(' ')[1] <= '16:00:00' else 'Buy' if idx in valleys and df.iloc[idx]['Local time'].split(' ')[1] >= '10:15:00' and df.iloc[idx]['Local time'].split(' ')[1] <= '15:00:00' else 'Hold' for idx in range(len(df))]
    df['Local time'] = pd.to_datetime(df['Local time'])
    df = df.set_index('Local time')
    df.to_csv(f'./daily_stock_data/{file}')
    dates = [idx.isoformat().replace(':', '-') for idx in df.index]
    
    for i in range(len(df)):
        if i %1000 == 0:
            print(i)
        if i-44 >= 0:
            date = df.iloc[i:i+1].index[0].isoformat().replace(':', '-')
            check_folder_exists(f"./stock_graphs/{file.split('.')[0]}/")
            image_file = f"./stock_graphs/{file.split('.')[0]}/{symbol}_{dates[i]}.png"
            if path.exists(image_file):
                continue
                
            mc = fplt.make_marketcolors(
                up='tab:blue',down='tab:red',
                edge='black',
                wick={'up':'blue','down':'red'},
                volume='green', alpha=1.0
            )
            s  = fplt.make_mpf_style(marketcolors=mc, mavcolors=["yellow","orange",'brown'])

            fig = fplt.plot(
                df.iloc[i-44:i+1],
                type='candle',
                style=s,
                volume=True,
                axisoff=True,
                mav=(5,8,13),
                figsize=(3.2,3.2),
                xlim=(12.5, 44.5),
                savefig=image_file,
                tight_layout=True
            )
```
